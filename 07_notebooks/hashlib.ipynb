## O que seu código já faz bem

- Percorre o dicionário

- Considera apenas valores str

```python
import hashlib

def add_hash(data: dict) -> str:
    parts = []

    for k in sorted(data.keys()):
        v = data[k]
        if isinstance(v, str):
            parts.append(v.strip().upper())

    raw = "|".join(parts)
    return hashlib.md5(raw.encode("utf-8")).hexdigest()
```

Incluir chave + valor no hash

```python
def add_hash(data: dict) -> str:
    parts = []

    for k in sorted(data.keys()):
        v = data[k]
        if v is not None:
            parts.append(f"{k}={str(v).strip().upper()}")

    raw = "|".join(parts)
    return hashlib.md5(raw.encode("utf-8")).hexdigest()
```

```python
registro = {
    "nome": "Luciana",
    "cidade": "São Paulo",
    "idade": 39
}

print(add_hash(registro))
```

## Usando como hash_diff (padrão Medallion)

```python
def add_hash(data: dict, ignore_keys=None) -> str:
    ignore_keys = ignore_keys or []
    parts = []

    for k in sorted(data.keys()):
        if k in ignore_keys:
            continue

        v = data[k]
        if v is not None:
            parts.append(f"{k}={str(v).strip().upper()}")

    raw = "|".join(parts)
    return hashlib.md5(raw.encode("utf-8")).hexdigest()
```

```python
registro = {
    "id": 10,
    "nome": "Luciana",
    "updated_at": "2026-01-01"
}

print(add_hash(registro, ignore_keys=["id", "updated_at"]))
```
